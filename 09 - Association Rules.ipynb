Data Preprocessing not required as it is already in transaction format.

#Apriori Algorithm

#1) Association rules with 10% Support and 70% confidence

```python
# with 10% Support

frequent_itemsets = apriori(book, min_support= 0.1, use_colnames=True)
frequent_itemsets
```

```python
# with 70% Confidence

rules = association_rules(frequent_itemsets, metric='lift', min_threshold=0.7)
rules
```



*   A leverage value of 0 indicates independence.

*  Range will be A high conviction value means that the consequent is highly
depending on the antecedent and range [0 inf]

```python
rules.sort_values('lift',ascending=False)
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

# Apriori Algorithm
1. Association rules with 10% Support and 70% confidence

```python
# with 10% Support
frequent_itemsets=apriori(movie2,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
#With 70% Confidence

rules = association_rules(frequent_itemsets,metric='lift',min_threshold=True)
rules
```

```python
# Using Lift Ration Rule
rules[rules.lift>1]
```

```python
# Visualization of obtained Rules
plt.scatter(rules['support'], rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

2. Association rules with 5% Support and 80% confidence

```python
# with 5% support

frequent_itemsets2 = apriori(movie2, min_support=0.05, use_colnames = True)
frequent_itemsets2
```

```python
# With 80% Confidence

rules2 = association_rules(frequent_itemsets2, metric='lift', min_threshold=0.8)
rules2
```

```python
#Lift Ratio >1

rules2[rules2.lift>1]
```
